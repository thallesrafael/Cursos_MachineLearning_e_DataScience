```python
def plotar_linha(titulo, xlabel, ylabel, x, y, dataset, hue=None):
    ax = sns.lineplot(x=x, y=y, hue=hue, data=dataset)
    ax.figure.set_size_inches(12,6)
    ax.set_title(titulo, loc='left', fontsize=20)
    ax.set_xlabel(xlabel, fontsize=15)
    ax.set_ylabel(ylabel, fontsize=15, rotation=90)
```

```python
def plotar_barra(titulo, xlabel, ylabel, x, y, dataset):
    plt.figure(figsize=(20,6))
    ax = plt.bar(dataset[x], dataset[y])
    plt.title(titulo, loc='left', fontsize=20)
    plt.xlabel(xlabel, fontsize=15)
    plt.ylabel(ylabel, fontsize=15, rotation=90)
```

```python
plotar_barra("Novos casos em SP", 'Data', 'Novos casos', 'data', 'novosCasos', df_sp)
```

```python
df_sp['aceleracaoCasos'] = df_sp['novosCasos'].diff()
df_sp['aceleracaoObitos'] = df_sp['novosObitos'].diff()
df_sp.head()
```

```python
plotar_barra('Aceleracao de bitos em SP', 'Data', 'Aceleracao', 'data', 'aceleracaoObitos', df_sp)
```

```python
# nao mostra os 7 primeiros dados em um e os 14 primeiros em outro
df_sp['mediaObitos'] = df_sp.novosObitos.rolling(window=7, center=False).mean()
df_sp['mediaObitos_14'] = df_sp.novosObitos.rolling(window=14, center=False).mean()
df_sp.head(10)
```

```python
plotar_linha('Mdia mvel de bitos em SP', 'Data', 'Mdia', 'data', 'mediaObitos', df_sp, None)
plotar_linha('Mdia mvel de bitos em SP', 'Data', 'Mdia', 'data', 'mediaObitos_14', df_sp, None)
plt.bar(df_sp['data'], df_sp['novosObitos'], color='lightgrey')
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
# Mostra a autocorrelao em funo dos lags
# Percebe-se que at cerca de 20 lags (dias, nesse caso) h uma correlao muito forte
# Essa correlao vai diminuindo com o tempo (isso se d porque a taxa de crescimento de bitos aumentou bastante)
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlao dos bitos', fontsize = 18, x = 0.26, y = 0.95)
autocorrelation_plot(df_sp['Obitos'])
ax = ax
```

```python
# O comportamento da sazonalidade se repete aqui para a autoccorelao dos novos casos
# Em lags multiplos de 7, a correlao  maior, como pode ser visto
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlao dos novos bitos', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(df_sp['novosObitos'])
ax=ax
```

```python
# Fazendo o grfico de autocorrelao para a mdia mvel de 7 dias
ax = plt.figure(figsize=(12, 6))
ax.suptitle('Correlao dos novos bitos (mdia mvel)', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(df_sp['mediaObitos'][7:]) # s h dados a partir do stimo dia por conta do 'rolling()'
ax = ax
```

```python
ax = plt.figure(figsize=(12, 6))
ax.suptitle('Autocorrelao da acelerao de bitos', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(df_sp['aceleracaoObitos'][1:])
ax = ax
```

```python
df_sp['mediaAccObitos'] = df_sp['aceleracaoObitos'][1:].rolling(window=7,center=False).mean()
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Autocorrelao da mdia da acelerao de bitos',fontsize=18,x=0.26,y=0.95)
autocorrelation_plot(df_sp['mediaAccObitos'][7:])
ax=ax
```

```python
# Colocando o dia da semana para cada dado para estudarmos melhor a sazonalidade
df_sp['mes'] = pd.to_datetime(df_sp['data'])
df_sp['dia_da_semana'] = df_sp['mes'].dt.day_name()
df_sp.head()
```

```python
dias_traduzidos = {'Monday': 'Segunda', 'Tuesday': "Tera", "Wednesday": 'Quarta', 
                   'Thursday': 'Quinta', 'Friday': 'Sexta', 'Saturday': 'Sabado',
                   'Sunday': 'Domingo'}
```

```python
def criar_features(df, label=None):
    df = df.copy()
    df['diasemana'] = df.mes.dt.weekday
    df['diames'] = df.mes.dt.day
    df['mes'] = df.mes.dt.month


    X = df[['diasemana', 'mes', 'diames']]

    if label:
        y= df[label]
        return X, y
    return X
```

```python
X, y = criar_features(df_sp, label='Obitos')
```

```python
df_final = pd.concat([X, y], axis=1)
```

```python
df_final.head(10)
```

```python
g = sns.pairplot(df_final.dropna(),
                hue='mes',
                x_vars=['diasemana', 'diames', 'mes'],
                y_vars='Obitos',
                palette = 'tab10')

g.fig.set_size_inches(12,6)
g.axes[0,0].set_xlabel('Dia da semana', fontsize=15)
g.axes[0,1].set_xlabel('Dia do ms', fontsize=15)
g.axes[0,1].set_xlabel('Ms', fontsize=15)
g.axes[0,1].set_ylabel('bitos', fontsize=15)

plt.suptitle('bitos por dia da semana, dia do ms e por ms', fontsize=15)
plt.show()
```
