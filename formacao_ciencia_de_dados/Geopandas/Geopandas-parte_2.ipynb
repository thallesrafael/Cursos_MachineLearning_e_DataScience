```python
import os
import zipfile
import shutil
```

```python
dir = '../Dados/RJ-SETOR/01.Dados/Mapas/RJ-SETOR/TEMP'
if not os.path.exists(dir):
    os.makedirs(dir)
```

```python
for root, dirs, files in os.walk('../Dados/RJ-SETOR/01.Dados/Mapas/RJ-SETOR/DADOS'):
    filenames = files
    
filenames
```

```python
for item in filenames:
    zip_ref = zipfile.ZipFile('../Dados/RJ-SETOR/01.Dados/Mapas/RJ-SETOR/DADOS/' + item, 'r')
    zip_ref.extractall(dir)
    zip_ref.close()
```

# Agregando regiões geográficas

```python
setor.plot(color = 'white', edgecolor = 'black', figsize = (15, 8))
```

```python
setor.head()
```

```python
setor_m = setor[['NM_BAIRRO', 'geometry']]
setor_m = setor_m.to_crs({'init': 'epsg:4326'})
```

```python
bairro = setor_m.dissolve(by = 'NM_BAIRRO')
```

```python
bairro
```

```python
bairro.plot(color = 'white', edgecolor = 'black', figsize = (15, 8))
```

```python
setor_m = setor[['NM_BAIRRO', 'geometry', 'V002']]
setor_m = setor_m.to_crs({'init': 'epsg:4326'})
```

```python
bairro = setor_m.dissolve(by = 'NM_BAIRRO', aggfunc = 'sum')
```

```python
bairro
```

```python
bairro.V002.sum()
```

```python
setor.V002.sum()
```

```python
bairro = bairro.to_crs('+proj=utm +zone=23, +south +ellps=GRS80 +tougs=84=0,0,0,0,0,0,0 +units=km +no_defs')
bairro.crs
```

```python
bairro = bairro.reset_index()
bairro
```

```python
dir = '../Dados/RJ-SETOR/01.Dados/Mapas/RJ-BAIRRO'
if not os.path.exists(dir):
    os.makedirs(dir)
    
bairro.to_file(dir + '/RJ-BAIRRO.shp')
```

# Gerando Visualizações dos Dados

Ajuda Foluim: https://github.com/python-visualization/folium/tree/main/folium

Exemplos: https://github.com/python-visualization/folium/tree/main/examples

```python
import folium
```

```python
rj = gpd.read_file('../Dados/RJ-MUNIC/RJ-MUNIC.shp')
```

```python
crs = {'init': 'epsg:4326'}
```

```python
setor = setor.fillna(0)
```

```python
rj = rj.to_crs(crs)
```

```python
y = rj.centroid.y.iloc[0]
y
```

```python
x = rj.centroid.x.iloc[0]
x
```

## Mapa Camada Simples

```python
base = folium.Map([y, x], zoom_start = 11, tiles = 'OpenStreetMap')
base.choropleth(rj)
base
```

```python
base = folium.Map([y, x], zoom_start = 11, tiles = 'OpenStreetMap')
geojson_rj = folium.GeoJson(rj)
geojson_rj.add_child(folium.Popup(rj.NM_MUNICIP.iloc[0]))
geojson_rj.add_to(base)
base
```
