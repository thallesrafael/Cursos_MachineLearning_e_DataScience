```python
y = df_sp['Casos']
y.head()
```

```python
# importando a ferramenta ADF que testa a estacionariedade da serie
from statsmodels.tsa.stattools import adfuller 
```

```python
adfuller(y.dropna())
```

```python
print(f'p-valor: {adfuller(y.dropna())[1]}')
```

Visto o alto valor do p-valor (muito acima de 0,05), pode-se concluir que a srie no  estacionaria:

- H tendncia
- H sazonalidade

Mdia e varincia mudam ao longo do tempo (nesse caso, a srie est crescendo)

```python
# para plotar ACF (autocorrelao) e PACF (autocorrelao parcial)
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
def plotarACF(y1, y2, y3): # plotar a srie, a primeira diferenciao e a segunda diferencio
        fig = plt.figure(figsize=(10, 10))
        ax1 = fig.add_subplot(311)
        fig = plot_acf(y1, ax=ax1, title='Original')
        ax2 = fig.add_subplot(312)
        fig = plot_acf(y2, ax=ax2, title='1 Diferenciao')
        ax3 = fig.add_subplot(313)
        fig = plot_acf(y3, ax=ax3, title='2 Diferenciao')
```

```python
plotarACF(y, y.diff().dropna(), y.diff().diff().dropna())
```

```python
plotar(y, y.diff().dropna(), y.diff().diff().dropna(), 'red', 'blue', 'green', 'Original', '1 Diferenciao', '2 Diferenciao')
```

Percebe-se que os comportamentos da primeira e da segunda diferenciao esto muito similares. Ento, talvez, a primeira diferenciao seja suficiente para tornar a srie estacionria. Aplicando a segunda diferenciao, estamos sobrediferenciando (diferenciando mais do que necessrio)

```python
# Plotando a autocorelao parcial
plot_pacf(y.diff().dropna(), lags=40); # dois pontos para aparesentar um unico resultado
```

```python
# Como est aparecendo uma forte autocorrelao negativa, indica que manipulamos demais os dados diferenciando duas vezes
# Inclusive est ultrapassando a parte azul clara, que indica o grau de significancia
plot_pacf(y.diff().diff().dropna(), lags=40);
```

Como est aparecendo uma forte autocorrelao negativa, indica que manipulamos demais os dados diferenciando duas vezes. Inclusive est ultrapassando a parte azul clara, que indica o grau de significancia

Essa sobrediferenciao pde prejudicar o modelo

Portanto, opta-se por utilizar a primeira diferenciao

## Plota-se o PACF e conta-se quantos pontos ficam fora da parte pintada (significncia). Isso vai definir o parmetro "q" do modelo MA (que faz uma regresso linear com os erros da previso - resduos). 

### Assim, tem que considerar quantos 'lags' esto adicionando essa informao:
   - Tem-se os dois primeiros pontos muito "fora"
   - Mas pode ser conservador e considerar um ponto s (o segundo). Assim, q=1

```python
plot_pacf(y.diff().dropna(), lags=20);
```

- <b>ARIMA(p=1, d=2, q=1) :</b> Utilizar a segunda diferenciao e lidar com o problema das aucorrelaes negativas (sobrediferenciao)

### Mtricas mais importantes so a AIC e BIC:
- <b>AIC:</b> Mtrica de distncia. Compara o resultado que obteve e o modelo estatstico real. Ou seja, quanto menor o valor de AIC, melhor
- <b>AIB:</b> Estimativa da funo de probabilidade do modelo. Ou seja, a probabilidade do modelo ser verdadeiro. Por tambms ser uma medida de distncia quanto menos o AIB, melhor.

 <b><u>Vale ressaltar que valores muito baixos tambm podem indicar overfitting, ou seja, modelo ter bom desempenho na base de treinamento, mas quando se deparar com dados que nunca viu, ter um desempenho ruim</u></b>

```python
import itertools
p = d = q = range(0,2)
pdq = list(itertools.product(p, d, q))
```

```python
pdq
```

```python
seasonal_pdq = [(x[0], x[1], x[2], 7) for x in list(itertools.product(p, d, q))]
```

```python
seasonal_pdq1 = [(x[0], x[1], x[2], 7) for x in pdq] #prefiro assim
```

```python
# Mostrando como ficaram os parametros com alguns exemplos
print('Exemplo de combinao de parmetros para o ARIMA Sazonal')
print(f'SARIMAX: {pdq[0]} x {seasonal_pdq1[0]}')
print(f'SARIMAX: {pdq[1]} x {seasonal_pdq1[1]}')
print(f'SARIMAX: {pdq[2]} x {seasonal_pdq1[2]}')
```

### Vamos calcular os modelos com todos os parmetros e usar mtricas para avali-los. Os com melhor desempenho sero os com parmetros vencedores

```python
y = df_sp['Casos']
y
```

```python
import statsmodels.api as sm

aic_lista = list()
for param in pdq:
    for param_seasonal in seasonal_pdq:
        mod = sm.tsa.statespace.SARIMAX(y, order=param,
                                       seasonal_order=param_seasonal)
        results = mod.fit()
        aic_lista.append(f'ARIMA {param} x {param_seasonal} - AIC {results.aic}')
```

```python
aic_lista
```

```python
mod = sm.tsa.statespace.SARIMAX(y, 
                                order = (1, 1, 1),
                               seasonal_order=(1, 1, 1, 7))
results = mod.fit()
```

```python
results.plot_diagnostics(figsize=(16,8))
plt.show
```
