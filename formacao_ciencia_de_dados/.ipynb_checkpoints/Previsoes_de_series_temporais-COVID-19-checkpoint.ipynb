```python
y = df_sp['Casos']
y.head()
```

```python
# importando a ferramenta ADF que testa a estacionariedade da serie
from statsmodels.tsa.stattools import adfuller 
```

```python
adfuller(y.dropna())
```

```python
print(f'p-valor: {adfuller(y.dropna())[1]}')
```

Visto o alto valor do p-valor (muito acima de 0,05), pode-se concluir que a série não é estacionaria:

- Há tendência
- Há sazonalidade

Média e variância mudam ao longo do tempo (nesse caso, a série está crescendo)

```python
# para plotar ACF (autocorrelação) e PACF (autocorrelação parcial)
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
def plotarACF(y1, y2, y3): # plotar a série, a primeira diferenciação e a segunda diferencição
        fig = plt.figure(figsize=(10, 10))
        ax1 = fig.add_subplot(311)
        fig = plot_acf(y1, ax=ax1, title='Original')
        ax2 = fig.add_subplot(312)
        fig = plot_acf(y2, ax=ax2, title='1ª Diferenciação')
        ax3 = fig.add_subplot(313)
        fig = plot_acf(y3, ax=ax3, title='2ª Diferenciação')
```

```python
plotarACF(y, y.diff().dropna(), y.diff().diff().dropna())
```

```python
plotar(y, y.diff().dropna(), y.diff().diff().dropna(), 'red', 'blue', 'green', 'Original', '1ª Diferenciação', '2ª Diferenciação')
```

Percebe-se que os comportamentos da primeira e da segunda diferenciação estão muito similares. Então, talvez, a primeira diferenciação seja suficiente para tornar a série estacionária. Aplicando a segunda diferenciação, estamos sobrediferenciando (diferenciando mais do que necessário)

```python
# Plotando a autocorelação parcial
plot_pacf(y.diff().dropna(), lags=40); # dois pontos para aparesentar um unico resultado
```

```python
# Como está aparecendo uma forte autocorrelação negativa, indica que manipulamos demais os dados diferenciando duas vezes
# Inclusive está ultrapassando a parte azul clara, que indica o grau de significancia
plot_pacf(y.diff().diff().dropna(), lags=40);
```

Como está aparecendo uma forte autocorrelação negativa, indica que manipulamos demais os dados diferenciando duas vezes. Inclusive está ultrapassando a parte azul clara, que indica o grau de significancia

Essa sobrediferenciação póde prejudicar o modelo

Portanto, opta-se por utilizar a primeira diferenciação

## Plota-se o PACF e conta-se quantos pontos ficam fora da parte pintada (significância). Isso vai definir o parâmetro "q" do modelo MA (que faz uma regressão linear com os erros da previsão - resíduos). 

### Assim, tem que considerar quantos 'lags' estão adicionando essa informação:
   - Tem-se os dois primeiros pontos muito "fora"
   - Mas pode ser conservador e considerar um ponto só (o segundo). Assim, q=1

```python
plot_pacf(y.diff().dropna(), lags=20);
```

# ARIMA(p, d = 1, q = 1)

```python
plot_acf(y.diff().dropna());
```

## Percebe-se, entretanto, que no ACF há muitos pontos fora da área de significância. 

### Deve-se levar em conta que o modelo ARIMA não tem bom desempenho quando há sazonalidade. O que há nesses dados mesmo após a primeira diferenciação. O modelo ARIMA sazonal teria melhor desempenho.

### Com isso, duas saídas:
- <b>ARIMA(p=1, d=1, q=1) :</b> Considerar o modelo com p=1 para deixá-lo mais simples (uma vez que mais complexidade não trará melhor desempenho uma vez que há o problema da sazonalidade) 
- <b>ARIMA(p=1, d=2, q=1) :</b> Utilizar a segunda diferenciação e lidar com o problema das aucorrelações negativas (sobrediferenciação)

```python
plot_acf(y.diff().diff().dropna());
```

### Partiremos do modelo mais simples para teste

```python
# Importando o modelo ARIMA
from statsmodels.tsa.arima_model import ARIMA
```

```python
modelo = ARIMA(y, order=(1,1,1))
modelo_fit = modelo.fit(disp=0)
print(modelo_fit.summary())
```

### Métricas mais importantes são a AIC e BIC:
- <b>AIC:</b> Métrica de distância. Compara o resultado que obteve e o modelo estatístico real. Ou seja, quanto menor o valor de AIC, melhor
- <b>AIB:</b> Estimativa da função de probabilidade do modelo. Ou seja, a probabilidade do modelo ser verdadeiro. Por tambéms ser uma medida de distância quanto menos o AIB, melhor.

 <b><u>Vale ressaltar que valores muito baixos também podem indicar overfitting, ou seja, modelo terá bom desempenho na base de treinamento, mas quando se deparar com dados que nunca viu, terá um desempenho ruim</u></b>

```python
import itertools
p = d = q = range(0,2)
pdq = list(itertools.product(p, d, q))
```

```python
pdq
```

```python
seasonal_pdq = [(x[0], x[1], x[2], 7) for x in list(itertools.product(p, d, q))]
```

```python
seasonal_pdq1 = [(x[0], x[1], x[2], 7) for x in pdq] #prefiro assim
```

```python
# Mostrando como ficaram os parametros com alguns exemplos
print('Exemplo de combinação de parâmetros para o ARIMA Sazonal')
print(f'SARIMAX: {pdq[0]} x {seasonal_pdq1[0]}')
print(f'SARIMAX: {pdq[1]} x {seasonal_pdq1[1]}')
print(f'SARIMAX: {pdq[2]} x {seasonal_pdq1[2]}')
```

### Vamos calcular os modelos com todos os parâmetros e usar métricas para avaliá-los. Os com melhor desempenho serão os com parâmetros vencedores

```python
y = df_sp['Casos']
y
```

```python
import statsmodels.api as sm

aic_lista = list()
for param in pdq:
    for param_seasonal in seasonal_pdq:
        mod = sm.tsa.statespace.SARIMAX(y, order=param,
                                       seasonal_order=param_seasonal)
        results = mod.fit()
        aic_lista.append(f'ARIMA {param} x {param_seasonal} - AIC {results.aic}')
```

```python
aic_lista
```

```python
mod = sm.tsa.statespace.SARIMAX(y, 
                                order = (1, 1, 1),
                               seasonal_order=(1, 1, 1, 7))
results = mod.fit()
```

```python
results.plot_diagnostics(figsize=(16,8))
plt.show
```
